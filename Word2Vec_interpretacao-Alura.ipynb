# One-hot encode
* A técnica de one-hot encoding é utilizada para representar variáveis categóricas como vetores binários.
* No contexto da vetorização de palavras, ela é usada para representar palavras como vetores, onde cada posição no vetor representa uma palavra única.
* No one-hot encoding, apenas uma posição no vetor possui o valor 1, enquanto todas as outras posições são 0.

```python

texto =[
        "tenha um bom dia",
        "tenha um péssimo dia",
        "tenha um ótimo dia",
        "tenha um dia ruim python java alura caelum papa"

]

vetorizador = CountVectorizer()
vetorizador.fit(texto)

```

```python
print(vetorizador.vocabulary_)
```

```python
vetor_bom = vetorizador.transform(["bom"])
print(vetor_bom.toarray())
```

```python
vetor_otimo = vetorizador.transform(["ótimo"])
print(vetor_otimo.toarray())
```

# Vetorização de texto

```python
artigo_treino.title.loc[12]
```

```python
def tokenizador(texto):
    texto = texto.lower()
    lista_alfanumerico = []

    for token_valido in nltk.word_tokenize(texto):
        if token_valido in string.punctuation:
          continue
        lista_alfanumerico.append(token_valido)

    return lista_alfanumerico

print("Texto Exemplo, 1234.".split())
tokenizador("Texto Exemplo, 1234.")
```

Palavras com poucas ocorrências são substituídas por UNKNOWN. Veja no [ artigo produzido](http://nilc.icmc.usp.br/nilc/index.php/repositorio-de-word-embeddings-do-nilc)

```python
def combinacao_de_vetores_por_soma(palavras_numeros):
    vetor_resultante = np.zeros(300)

    # pn eh palavras e numeros
    for pn in palavras_numeros:
        try:
            vetor_resultante += modelo.get_vector(pn)
        # Veja o artigo produzido
        except KeyError:
            # Se é um número, então transforma em 00...
            if pn.isnumeric():
              # 123 -> 000
               pn = "0"*len(pn)
               vetor_resultante += modelo.get_vector(pn)
            # Se é uma palavra com baixa frequência, então é unknown
            else:
                vetor_resultante += modelo.get_vector("unknown")

    return vetor_resultante
```

[Nesta thread](https://www.kaggle.com/c/quora-insincere-questions-classification/discussion/71778), você encontrará discussão e recomendação de diversas técnicas de combinação.

```python
palavras_numeros = tokenizador("texto exemplo caelumx")
vetor_texto = combinacao_de_vetores_por_soma(palavras_numeros)
print(len(vetor_texto))
print(vetor_texto)
```

```python
def matriz_vetores(textos):
    x = len(textos)
    y = 300
    matriz = np.zeros((x,y))

    for i in range(x):
        palavras_numeros = tokenizador(textos.iloc[i])
        matriz[i] = combinacao_de_vetores_por_soma(palavras_numeros)

    return matriz

matriz_vetores_treino = matriz_vetores(artigo_treino.title)
matriz_vetores_teste = matriz_vetores(artigo_teste.title)
print(matriz_vetores_treino.shape)
print(matriz_vetores_teste.shape)
```

# Classificação

**Log Reg**
* utilizado para problemas de classificação binária, onde o objetivo é prever a probabilidade de um evento ocorrer ou não

```python
lr = LogisticRegression(max_iter = 200)
lr.fit(matriz_vetores_treino, artigo_treino.category)
```

```python
lr.score(matriz_vetores_teste, artigo_teste.category)
```

```python
artigo_teste.category.unique()
```

**Precisão e Recall**
* A precisão mede a proporção de instâncias classificadas corretamente como positivas em relação ao total de instâncias classificadas como positivas.
* Já a revocação mede a proporção de instâncias classificadas corretamente como positivas em relação ao total de instâncias que realmente são positivas.

```python
label_prevista = lr.predict(matriz_vetores_teste)
cr = classification_report(artigo_teste.category, label_prevista)
print(cr)
```

## Classificando com Skip-gram
* utiliza uma janela deslizante para capturar o contexto das palavras e cria vetores de palavras que representam a semântica das palavras.

### Load do modelo

```python
!unzip "/content/drive/MyDrive/w2v-Alura/skip_s300.zip"
```

```python
modelo_skip = KeyedVectors.load_word2vec_format("skip_s300.txt")
```

### Ajustando as funções

```python
def combinacao_de_vetores_por_soma_skip(palavras_numeros):
    vetor_resultante = np.zeros(300)

    # pn eh palavras e numeros
    for pn in palavras_numeros:
        try:
            vetor_resultante += modelo_skip.get_vector(pn)
        # Veja o artigo produzido
        except KeyError:
            # Se é um número, então transforma em 00...
            if pn.isnumeric():
              # 123 -> 000
               pn = "0"*len(pn)
               vetor_resultante += modelo_skip.get_vector(pn)
            # Se é uma palavra com baixa frequência, então é unknown
            else:
                vetor_resultante += modelo_skip.get_vector("unknown")

    return vetor_resultante

def matriz_vetores_skip(textos):
    x = len(textos)
    y = 300
    matriz = np.zeros((x,y))

    for i in range(x):
        palavras_numeros = tokenizador(textos.iloc[i])
        matriz[i] = combinacao_de_vetores_por_soma_skip(palavras_numeros)

    return matriz

matriz_vetores_treino = matriz_vetores(artigo_treino.title)
matriz_vetores_teste = matriz_vetores(artigo_teste.title)
print(matriz_vetores_treino.shape)
print(matriz_vetores_teste.shape)
```

### Matriz de treino e teste

```python
matriz_vetores_treino_skip = matriz_vetores_skip(artigo_treino.title)
matriz_vetores_teste_skip = matriz_vetores_skip(artigo_teste.title)
```

### Previsão

```python
lr_skip = LogisticRegression(max_iter=1000)
lr_skip.fit(matriz_vetores_treino_skip, artigo_treino.category)
```

```python
label_prevista_skip = lr_skip.predict(matriz_vetores_teste_skip)
cr_skip = classification_report(artigo_teste.category, label_prevista_skip)
print(cr_skip)
```

# Diferença entre Cbow e SkipGram
* No CBOW, o modelo tenta prever a palavra de destino com base em um contexto de palavras de entrada.

* Ele utiliza a média dos vetores de palavras de entrada para obter uma representação do contexto e, em seguida, tenta prever a palavra de destino.

* O CBOW é mais rápido de treinar e tende a funcionar bem em textos menores.


---


* Já o Skipgram, ao contrário do CBOW, tenta prever as palavras de contexto com base em uma palavra de entrada.

* Ele desliza uma janela de contexto em torno da palavra de entrada e tenta prever as palavras vizinhas.

* O Skipgram é mais lento de treinar, mas tende a funcionar melhor em textos maiores e é considerado um pouco melhor em identificar a semântica das palavras.

---

* Em resumo, o CBOW tenta prever a palavra de destino com base no contexto, enquanto o Skipgram tenta prever as palavras de contexto com base na palavra de entrada.
